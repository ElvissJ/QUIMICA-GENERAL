```python
import pandas as pd
import math
# Datos
Temperaturas = [100, 300, 700]
Constante_Boltzmann = 1.38e-23
Masa_N2 = 28
Numero_Avogadro = 6.0221e23
Presion = 1.0
Volumen = 1.0

data = []
for Temp in Temperaturas:
    n = (Presion * Volumen) / (8.314 * Temp)
    Numero_moleculas = n * Numero_Avogadro

    Velocidad = math.sqrt((3 * Constante_Boltzmann * Temp) / (Masa_N2 * 1.60218e-27))

    data.append([Temp, Velocidad, Numero_moleculas])

df = pd.DataFrame(data, columns=['Temperatura (K)','Velocidad (m/s)','Cantidad de Mol√©culas de N2'])

# Mostrar la tabla en Colab
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
D_e = ['100 K', '300 k', '700 k']
D_s = [303.784424, 526.170056, 803.738037]
plt.bar(D_e, D_s, color = "orange")
plt.xticks(rotation=90)
plt.show();
```

Se observa que a medida que se aumenta la temperatura, m√°s mol√©culas se mueven a mayores velocidades. Es decir son directamente proporcionales. Adem√°s se evidencia que la media de la velocidad es la que hallamos mediante la f√≥rmula usando la constante de Boltzmann. En el gr√°fico la media es hallada al sumar las velocidades m√≠nimas y las m√°ximas para cada temperatura, y dividirlas por el n√∫mero de p√°rticulas. Gr√°ficamente ser√≠a como partir las curvas en dos partes iguales, y el valor del que sale la l√≠nea que las divide es la media. La mediana tiene el mismo valor que la media, ya que no importa el orden con el que se midan las part√≠culas, pues estas son iguales. La distribuci√≥n es sim√©trica. En cu√°nto a la moda, es la velocidad que adquiere el mayor n√∫mero de mol√©culas. gr√°ficamente ser√≠a el valor del que sale la l√≠nea que pasa por el valor m√°ximo del n√∫mero de part√≠culas de la gr√°fica.

```python
import pandas as pd
import math
# Datos
Masas = [70.90, 28.02, 4.003]
Constante_Boltzmann = 1.38e-23
Temperatura = 300
Numero_Avogadro = 6.0221e23
Presion = 1.0
Volumen = 1.0

data = []
for Mass in Masas:
    n = (Presion * Volumen) / (8.314 * 300)
    Numero_moleculas = n * Numero_Avogadro

    Velocidad = math.sqrt((3 * Constante_Boltzmann * 300) / (Mass * 1.60218e-27))

    data.append([Mass, Velocidad, Numero_moleculas])

df = pd.DataFrame(data, columns=['Masa molecular (g/mol)','Velocidad (m/s)','Cantidad de Mol√©culas'])

# Mostrar la tabla en Colab
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
D_e = ['70.90 g/mol', '28.02 g/mol', '4.003 g/mol']
D_s = [330.660279, 525.982239, 1391.593366]
plt.bar(D_e, D_s, color = "purple")
plt.xticks(rotation=90)
plt.show();
```

Adem√°s se evidencia que la media de la velocidad es la que hallamos mediante la f√≥rmula usando la constante de Boltzmann. Gr√°ficamente ser√≠a como partir las curvas en dos partes iguales, y el valor del que sale la l√≠nea que las divide es la media. La mediana tiene el mismo valor que la media, ya que no importa el orden con el que se midan las part√≠culas, pues estas son iguales. La distribuci√≥n es sim√©trica. En cu√°nto a la moda, es la velocidad que adquiere el mayor n√∫mero de mol√©culas. gr√°ficamente ser√≠a el valor del que sale la l√≠nea que pasa por el valor m√°ximo del n√∫mero de part√≠culas de la gr√°fica.
